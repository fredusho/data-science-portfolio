# Outliers

```python
outliers_iqr = {}

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    outliers_iqr[col] = {
        'count': outliers.shape[0],
        'lower_bound': lower_bound,
        'upper_bound': upper_bound
    }

outliers_iqr
```

```python
def procesar_df(df):
    df = df.copy()  # para no modificar el df original

    # --- LIMPIEZA DE COLUMNAS NUMÉRICAS ---

    # User Rating: pasar a float
    df['User Rating'] = pd.to_numeric(df['User Rating'], errors='coerce')

    # Number of Votes: eliminar comas y convertir a float
    df['Number of Votes'] = (
        df['Number of Votes']
        .astype(str)
        .str.replace(',', '', regex=False)
        .replace('nan', np.nan)
        .astype(float)
    )

    # Runtime: extraer número desde strings como '24 min'
    df['Runtime'] = (
        df['Runtime']
        .astype(str)
        .str.extract(r'(\d+)', expand=False)
        .astype(float)
    )

    # Gross: eliminar comas y convertir
    df['Gross'] = (
        df['Gross']
        .astype(str)
        .str.replace(',', '', regex=False)
        .replace('nan', np.nan)
        .astype(float)
    )

    # --- TRANSFORMACIONES ---

    # Agregar columnas logarítmicas
    df['Number of Votes Log'] = np.log1p(df['Number of Votes'])
    df['Gross Log'] = np.log1p(df['Gross'])

    # Eliminar columnas no necesarias si existen
    columnas_a_eliminar = ['Metascore', 'Year', 'Certificate',
                           'Summary', 'Stars', 'Episode Title']
    columnas_presentes = [col for col in columnas_a_eliminar if col in df.columns]
    df = df.drop(columns=columnas_presentes)

    # Seleccionar columnas finales útiles si están presentes
    columnas_finales = ['Title', 'Genre', 'User Rating', 'Number of Votes',
                        'Runtime', 'Gross', 'Episode',
                        'Number of Votes Log', 'Gross Log']
    columnas_finales = [col for col in columnas_finales if col in df.columns]

    return df[columnas_finales]
```

# Foward Selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import numpy as np

def forward_selection(df, features, target, cv=5):
    # 1. Filtrar filas donde target sea NaN
    df_clean = df[features + [target]].dropna(subset=[target]).copy()

    remaining = features.copy()
    selected = []
    best_score = -np.inf

    # 2. Iterar hasta que no mejore el score
    while remaining:
        scores = {}
        for feat in remaining:
            X = df_clean[selected + [feat]].fillna(0)
            y = df_clean[target]
            # Validación cruzada con R²
            scores[feat] = cross_val_score(
                LinearRegression(), X, y, cv=cv, scoring='r2'
            ).mean()

        # 3. Elegir la mejor variable de esta ronda
        best_feat, best_feat_score = max(scores.items(), key=lambda x: x[1])
        if best_feat_score > best_score:
            selected.append(best_feat)
            remaining.remove(best_feat)
            best_score = best_feat_score
        else:
            break

    return selected

# --- Uso ---
# Asegúrate de haber creado antes:
# df['Number of Votes Log'] = np.log1p(df['Number of Votes'])
# df['Gross Log']           = np.log1p(df['Gross'])

candidates = ['Number of Votes', 'Runtime', 'Gross', 'Episode',
              'Number of Votes Log', 'Gross Log']
selected_forward = forward_selection(df, candidates, 'User Rating')
print("Seleccionadas por Forward Selection:", selected_forward)
```

Ambos métodos coinciden en que las variables más predictivas de la calificación de usuario son Number of Votes Log y Gross Log, en línea con lo observado en el análisis exploratorio. runtime y episode no aportan información relevante para User Rating.
