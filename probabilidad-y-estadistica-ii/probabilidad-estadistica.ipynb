```python
import pandas as pd
import numpy as np
data = pd.read_csv('/content/worldcup2014.csv')
data.head()
data
```

```python
# Total de equipos en el dataset
total_teams = len(data)
total_teams
```

```python
# Crear una función para calcular probabilidades
def calculate_probability(condition):
    count = len(data[condition])
    return count / total_teams
```

```python
# 1f: Probabilidad de clasificar a la segunda ronda
prob_clasificar = calculate_probability(data['clasificado'] == 1)

{
    "prob_clasificar": prob_clasificar,
}
```

```python
# 1g: Probabilidad de ganar al menos un partido
prob_ganar_un_partido = calculate_probability(data['juegos_ganados'] > 0)

{
    "prob_ganar_un_partido": prob_ganar_un_partido
}
```

2 Parte, A partir de las probabilidades anteriores, y sin realizar conteos, calcular las
siguientes probabilidades:
a. pertenecer a Europa, si se sabe que clasificó
b. pertenecer a Sudamérica, si se sabe que clasificó
c. pertenecer a África, si se sabe que ganó al menos un partido
d. haber ganado al menos un partido, si se sabe que clasificó
e. haber ganado al menos un partido, si se sabe que pertenece a Norteamérica
f. NO haber clasificado, si se sabe que pertenece a Á

```python
# creamos primero una función para calcular probabilidades condicionales
def calculate_conditional_probability(condition_A, condition_B):
    count_B = len(data[condition_B])
    if count_B == 0:  # Evitar división por cero
        return 0
    count_A_and_B = len(data[condition_A & condition_B])
    return count_A_and_B / count_B

# 2a-b: Probabilidad de pertenecer a un continente dado que clasificó
prob_europe_given_clasified = calculate_conditional_probability(
    data['continent'] == 'europe', data['clasificado'] == 1
)
prob_southamerica_given_clasified = calculate_conditional_probability(
    data['continent'] == 'southamerica', data['clasificado'] == 1
)

# 2c: Probabilidad de pertenecer a África dado que ganó al menos un partido
prob_africa_given_win = calculate_conditional_probability(
    data['continent'] == 'africa', data['juegos_ganados'] > 0
)

# 2d: Probabilidad de ganar al menos un partido dado que clasificó
prob_win_given_clasified = calculate_conditional_probability(
    data['juegos_ganados'] > 0, data['clasificado'] == 1
)

# 2e: Probabilidad de ganar al menos un partido dado que pertenece a Norteamérica
prob_win_given_northamerica = calculate_conditional_probability(
    data['juegos_ganados'] > 0, data['continent'] == 'northamerica'
)

# 2f: Probabilidad de NO clasificar dado que pertenece a África
prob_not_clasified_given_africa = calculate_conditional_probability(
    data['clasificado'] == 0, data['continent'] == 'africa'
)

# Resultados de las probabilidades condicionales
{
    "prob_europe_given_clasified": prob_europe_given_clasified,
    "prob_southamerica_given_clasified": prob_southamerica_given_clasified,
    "prob_africa_given_win": prob_africa_given_win,
    "prob_win_given_clasified": prob_win_given_clasified,
    "prob_win_given_northamerica": prob_win_given_northamerica,
    "prob_not_clasified_given_africa": prob_not_clasified_given_africa
}
```
