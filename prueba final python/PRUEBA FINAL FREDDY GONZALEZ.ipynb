```python
# Cargar las tablas en dataframes
orders_df = pd.read_sql("SELECT * FROM orders", engine)
orders_df
```

2. Realiza el cruce entre los DataFrames, asegurándote de utilizar correctamente el
parámetro validate para asegurar la integridad referencial.

```python
# Combinamos primero Orders y Cuastomers uno a muchos ya que un cliente puede tener varias ordenes 
orders_customers_df = pd.merge(orders_df, customers_df, on="customerNumber", how="inner", validate="many_to_one")
orders_customers_df
```

```python
# Unimos el resultado anterior (orders_customers_df) con orderdetails a través de orderNumber
orders_customers_details_df = orders_customers_df.merge(orderdetails_df, on="orderNumber", validate="one_to_many")
orders_customers_details_df
```

```python
# Unimos el resultado anterior (orders_customers_details_df) con products a través de productCode
orders_customers_details_products_df = orders_customers_details_df.merge(products_df, on="productCode", validate="many_to_one")
orders_customers_details_products_df
```

```python
# Unimos el resultado anterior (orders_customers_details_products_df) con employees a través de salesRepEmployeeNumber
final_df = orders_customers_details_products_df.merge(employees_df, left_on="salesRepEmployeeNumber", right_on="employeeNumber", validate="many_to_one")
final_df
```

3. Agrega las siguientes columnas, considerando su nombre y la fórmula asociada- venta: quantityOrdered*priceEach- costo: quantityOrdered*buyPrice- ganancia: considerando las columnas anteriores

```python
# Agregar la columna 'venta'
final_df['venta'] = final_df['quantityOrdered'] * final_df['priceEach']

# Agregar la columna 'costo'
final_df['costo'] = final_df['quantityOrdered'] * final_df['buyPrice']

# Agregar la columna 'ganancia' como la diferencia entre 'venta' y 'costo'
final_df['ganancia'] = final_df['venta'] - final_df['costo']

# Visualizar las primeras filas del dataframe con las nuevas columnas
final_df[['quantityOrdered', 'priceEach', 'buyPrice', 'venta', 'costo', 'ganancia']].head()
```

4. ¿Cuál fue el total de ventas por línea de productos? Incluye una fila de totales.

```python
ventas_por_linea = final_df.groupby('productLine')['venta'].sum().reset_index()

# Renombrar la columna para mayor claridad
ventas_por_linea.rename(columns={'venta': 'total_venta'}, inplace=True)

# Agregar una fila de totales
total_ventas = pd.DataFrame({'productLine': ['Total'], 'total_venta': [ventas_por_linea['total_venta'].sum()]})
ventas_por_linea = pd.concat([ventas_por_linea, total_ventas], ignore_index=True)

# Visualizar el resultado
ventas_por_linea
```
