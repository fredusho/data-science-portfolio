a. carga de datos .pkl

```python
import pandas as pd

# Cargar los archivos .pkl y convertirlos en DataFrames
officers_df = pd.read_pickle('/content/officers.pkl')
incidents_df = pd.read_pickle('/content/incidents.pkl')
subjects_df = pd.read_pickle('/content/subjects.pkl')

# Mostrar las primeras filas de cada DataFrame para verificar
officers_df.head()
```

b. Genera una tabla que contenga la unión de las 3 tablas. hint: utiliza sufijos para para las columnas que se llaman igual usando el parámetro suffixes de pd.merge().

```python
# Unir 'officers_df' con 'subjects_df' en base a la columna 'case_number'
officers_subjects_merged = pd.merge(officers_df, subjects_df, on='case_number', suffixes=('_officer', '_subject'))

# Unir el resultado con 'incidents_df' también en base a la columna 'case_number'
final_merged_df = pd.merge(officers_subjects_merged, incidents_df, on='case_number', suffixes=('_merged', '_incident'))

# Mostrar las primeras filas de la tabla unida
final_merged_df.head()
```

c. Verifica si hay filas duplicadas; si es así, elimínalas.

```python
# Contar los valores en la columna 'gender_subject' y filtrar por el género 'F'
female_subjects_count = final_merged_df['gender_subject'].value_counts().get('F', 0)

# Mostrar el resultado
print(f'Número de sujetos de género F: {female_subjects_count}')
```

e. ¿En cuántos números de caso hay por lo menos una sospechosa que sea mujer? hint: utiliza el método unique() para obtener los valores únicos de una columna específica de un DataFrame luego de filtrar.

```python
# Filtrar los casos donde el género de los sujetos es 'F' (femenino)
female_subjects_cases = final_merged_df[final_merged_df['gender_subject'] == 'F']

# Obtener los valores únicos de 'case_number' en estos casos
unique_female_cases = female_subjects_cases['case_number'].unique()

# Contar el número de casos únicos
female_cases_count = len(unique_female_cases)

# Mostrar el resultado
print(f'Número de casos con al menos una sospechosa mujer: {female_cases_count}')
```

f. Genera una tabla pivote que muestre en las filas el género del oficial y en las columnas el género del subject. ¿Cómo interpretas los valores que muestra esta vista?

```python
# Crear una tabla pivote que cuente las ocurrencias de las combinaciones de género del oficial y del sujeto
pivot_table = pd.pivot_table(final_merged_df,
                             index='gender_officer',
                             columns='gender_subject',
                             aggfunc='size',
                             fill_value=0)

# Mostrar la tabla pivote
pivot_table
```

Los valores en la tabla pivot anterior representan el número de incidentes en los que un oficial de un género específico interactuó con un sujeto de otro género

a. Carga los datos y crea un DataFrame con ellos

```python
cleaned_ds_jobs_df = pd.read_csv('/content/Cleaned_DS_Jobs.csv')

# Mostrar las primeras filas del DataFrame para verificar
cleaned_ds_jobs_df.head()
```

b. Utiliza la siguiente lista de valores que serán considerados como nulos: ["na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"] (hint: utiliza el método replace para reemplazar los valores indicados por np.nan)

```python
import numpy as np

# Lista de valores a considerar como nulos
valores_nulos = ["na", "NA", -1, "0", "-1", "null", "n/a", "N/A", "NULL"]

# Reemplazar los valores indicados por pd.NA
cleaned_ds_jobs_df.replace(valores_nulos, pd.NA, inplace=True)

# Mostrar las primeras filas del DataFrame para verificar el cambio
print("DataFrame después de reemplazar valores nulos:")
cleaned_ds_jobs_df.head()
```

c. Elimina todas las filas con datos faltantes. (hint: utiliza el método .dropna())

```python
# Eliminar todas las filas con datos faltantes
cleaned_ds_jobs_df.dropna(inplace=True)

# Mostrar las primeras filas del DataFrame después de eliminar filas con datos faltantes
cleaned_ds_jobs_df.head()
```

d. A partir de la columna “Salary Estimate”, genera dos columnas: Salario Estimado Mínimo y Máximo. (hint: Utiliza el método apply sobre la columna.)

```python
# Función para separar el salario mínimo y máximo
def split_salary(salary):
    min_salary, max_salary = salary.split('-')  # Separar por el guion
    return pd.Series([int(min_salary), int(max_salary)])  # Retornar como dos columnas

# Aplicar la función y crear las columnas de salario mínimo y máximo
cleaned_ds_jobs_df[['Salario Mínimo', 'Salario Máximo']] = cleaned_ds_jobs_df['Salary Estimate'].apply(split_salary)

# Mostrar el DataFrame resultante
cleaned_ds_jobs_df
```

e. Realiza la recodificación de la columna Size con los valores de la siguiente tabla: (hint: utilice reemplazo con diccionario usando el método replace sobre la columna.)

```python
recodificacion_size = {
    '10000+ employees': 'Mega Empresas',
    '5001 to 10000 employees': 'Grandes Empresas',
    '1001 to 5000 employees': 'Medianas Empresas',
    '201 to 500 employees': 'Pequeñas Empresas',
    '51 to 200 employees': 'Pequeñas Grandes Empresas',
    '501 to 1000 employees': 'Microempresas',
    'Unknown': 'Empresas sin Información'
}

# Crear la nueva columna "descripcion" usando el diccionario
cleaned_ds_jobs_df['descripcion'] = cleaned_ds_jobs_df['Size'].replace(recodificacion_size)

# Mostrar las primeras filas del DataFrame para verificar el cambio
print("DataFrame después de agregar la columna 'descripcion':")
cleaned_ds_jobs_df
```

f. Genera una tabla pivote que muestre la media del salario estimado mínimo y la media del salario estimado máximo por tamaño de empresa. (hint: utiliza pd.pivot_table para generar la vista adecuada con las columnas generadas.)

```python
# Generar la tabla pivote
tabla_pivote = pd.pivot_table(
    cleaned_ds_jobs_df,
    values=['Salario Mínimo', 'Salario Máximo'],  # Columnas a promediar
    index='descripcion',  # Agrupar por tamaño de empresa
    aggfunc='mean'  # Calcular la media
)

# Mostrar la tabla pivote
print("Tabla pivote de la media del salario estimado mínimo y máximo por tamaño de empresa:")
tabla_pivote
```
